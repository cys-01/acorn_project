```python
df = weapon_ex_with_description

# 2) 용어별로 용례와 출처를 리스트로 묶기 (중복 제거, 순서 유지)
grouped = (
    df
    .groupby('용어')
    .agg({
        '용례': lambda s: list(dict.fromkeys(s)),   # 용례 리스트
        '출처': lambda s: list(dict.fromkeys(s))    # 출처 리스트
    })
    .reset_index()
    .rename(columns={
        '용례': '용례리스트',
        '출처': '출처리스트'
    })
)

# 3) 결과 확인
print(grouped.head())
```

# 최종 결합

```python
df1 = grouped   # 이미 리스트로 묶여 있는 용례리스트/출처리스트
df2 = grouped_2  # 개별 리스트가 아닌 ‘용례’, ‘출처’ 컬럼

# --- 2) df2도 같은 이름의 리스트 컬럼으로 바꿔 줍니다 ---
df2 = df2.rename(columns={
    '용례': '용례리스트_2',
    '출처': '출처리스트_2'
})

# 만약 여기서 df2['용례리스트_2'] 가 문자열로 된 리스트라면
# ast.literal_eval 등으로 실제 리스트로 변환해 두세요.

# --- 3) outer merge로 합칩니다 (양쪽 용어 모두 포함) ---
merged = pd.merge(
    df1,
    df2,
    on='용어',
    how='outer',
    suffixes=('_1','_2')
)

# --- 4) 두 리스트 컬럼을 합치고 중복 제거하는 헬퍼 함수 ---
def combine_lists(a, b):
    la = a if isinstance(a, list) else []
    lb = b if isinstance(b, list) else []
    # 순서 유지하면서 중복 제거
    return list(dict.fromkeys(la + lb))

# --- 5) 새로 합친 컬럼 생성 ---
merged['용례리스트']   = merged.apply(lambda r: combine_lists(
    r.get('용례리스트'), r.get('용례리스트_2')), axis=1)
merged['출처리스트']   = merged.apply(lambda r: combine_lists(
    r.get('출처리스트'), r.get('출처리스트_2')), axis=1)

# --- 6) 불필요 컬럼 정리 & 정렬 ---
final = merged[['용어','용례리스트','출처리스트']].sort_values('용어')
```

```python
final
```

```python
# --- 7) CSV로 저장 ---
final.to_csv(
    '/content/drive/MyDrive/Hanhwa/crawling_project/crawling_project/data/merged_grouped_final.csv',
    index=False,
    encoding='utf-8-sig'
)
print("✔ 최종 병합 결과 저장 완료")
```

```python
final['용례개수'] = final['용례리스트'].apply(len)

# 2) 총 용례 개수 계산
total_examples = final['용례개수'].sum()

# 3) 결과 출력
print(f"▶ 총 용례 개수: {total_examples}건\n")

print("▶ 용어별 용례 개수:")
print(final[['용어','용례개수']])
```
