# 텍스트 추출 후, 용례 매칭 함수

```python
import os
import re
import fitz  # PyMuPDF
import pandas as pd

# PDF 텍스트 추출 함수
def extract_text_from_pdf(pdf_path):
    doc = fitz.open(pdf_path)
    full_text = []
    for page in doc:
        text = page.get_text()
        full_text.append(text)
    doc.close()
    return "\n".join(full_text)

# 문장 단위로 나누는 함수
def split_sentences(text):
    sentences = re.split(r'(?<=[\.!?])\s+', text)
    return [s.replace('\n', '') for s in sentences if len(s) > 10] # 길이가 10 이상인 문장만 가져옴

# 여러 PDF 파일에서 텍스트 추출
def extract_all_sentences(pdf_folder):
    all_sentences = []
    for filename in os.listdir(pdf_folder):
        if filename.endswith(".pdf"):
            pdf_path = os.path.join(pdf_folder, filename)
            print(f"📄 처리 중: {filename}")  # 현재 진행 중인 PDF 출력
            text = extract_text_from_pdf(pdf_path)
            sentences = split_sentences(text)
            all_sentences.extend(sentences)
    return list(set(all_sentences))  # 중복 제거 => 다른 pdf상에 같은 문장이 있는 경우, 중복제거 => 하지만, 쉼표나 "및"과 같은 말이 추가되어 같은 문장으로 인식하지 못하지만 결국 거의 같은 문장은 제거 못함
    # ex: 1) 수상전력은 유도탄정, 어뢰정, 경비정, 화력 지원정 등 대부분 소형 고속함정으로 구성되어 있으며, 지상작전과 연계하여 지상군의 진출 지원과 연안 방어 등의 임무를 수행한다
    #     2) 수상전력은 유도탄정, 어뢰정, 소형경비정, 화력지원정 등 대부분 소형 고속함정으로 구성되어 있으며, 지상작전과 연계하여 지상군 진출을 지원하고 연안 방어 등의 임무를 수행한다.

# 용례 매핑 함수
def extract_examples_for_terms(sentences, term_list):
    map_dict = {term: [] for term in term_list}
    for sentence in sentences:
        for term in term_list:
            if term in sentence:
                map_dict[term].append(sentence)
    # 중복 문장 제거
    for term in map_dict:
        map_dict[term] = list(set(map_dict[term]))
    return map_dict

# CSV 저장 함수
def save_examples_to_csv(map_dict, output_path):
    df = pd.DataFrame([
        {"용어": term, "용례": examples}
        for term, examples in map_dict.items()
    ])
    df.to_csv(output_path, encoding="utf-8-sig", index=False)

# === 실행 ===
if __name__ == "__main__":
    # PDF 폴더 경로
    pdf_folder = "/content/drive/MyDrive/Hanhwa/crawling_project/crawling_project/data"

    # 무기 용어 리스트
    weapon = pd.read_csv("/content/drive/MyDrive/Hanhwa/crawling_project/crawling_project/data/무기의세계_불필요한것 제외.csv")
    term_list = list(weapon['용어'])

    # 전체 문장 수집
    all_sentences = extract_all_sentences(pdf_folder)

    # 용어별 용례 추출
    map_dict = extract_examples_for_terms(all_sentences, term_list)

    # CSV로 저장
    save_examples_to_csv(map_dict, "/content/drive/MyDrive/Hanhwa/crawling_project/crawling_project/data/weapon_examples.csv")

    print("✅ CSV 저장 완료: weapon_examples.csv")
```
