```python
def group_manual_csv(input_csv: str, output_csv: str):
    # 1) 기존 CSV 로드 (UTF-8 BOM 포함)
    df = pd.read_csv(input_csv, encoding='utf-8-sig')

    # 2) 컬럼 이름 확인 (필요시 조정)
    print("원본 컬럼:", df.columns.tolist())

    # 3) term별로 pdf와 example을 리스트로 묶기
    grouped = (
        df
        .groupby('용어')
        .agg({
            '출처': lambda s: list(dict.fromkeys(s)),   # PDF 출처 중복 제거 후 리스트
            '용례': list                            # 예시 문장 리스트
        })
        .reset_index()
        .rename(columns={
            '출처': '출처',
            '용례': '용례'
        })
    )

    # 4) 결과 확인
    print(grouped.head())

    # 5) CSV로 저장 (UTF-8 BOM 포함)
    grouped.to_csv(output_csv, index=False, encoding='utf-8-sig')
    print(f"✔ '{output_csv}'에 저장 완료!")

if __name__ == "__main__":
    input_csv = '/content/drive/MyDrive/Hanhwa/crawling_project/crawling_project/국방백서_국방연구소_결합_중복제거버전.csv'
    output_csv = "/content/drive/MyDrive/Hanhwa/crawling_project/crawling_project/data/국방백서_국방연구소_결합_중복제거버전_일대다.csv"
    group_manual_csv(input_csv, output_csv)
```
